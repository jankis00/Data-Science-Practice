```python
import pandas as pd
import numpy as np
import urllib3
import json
import time

URL_SERVER="https://la2.api.riotgames.com"
RIOT_KEY="RGAPI-f968c491-eaa2-427c-8ae6-c126f081fe43"
SUMMONER_NAME="jankis"
url= urllib3.PoolManager()
summoner_data= url.request('GET',URL_SERVER+'/lol/summoner/v4/summoners/by-name/'+SUMMONER_NAME,headers={'X-Riot-Token': RIOT_KEY})
json_summoner_data= json.loads(summoner_data.data)
account_id=json_summoner_data["accountId"]
Total_data=pd.DataFrame()
summoner_games_data= url.request('GET',URL_SERVER+'/lol/match/v4/matchlists/by-account/'+account_id,headers={'X-Riot-Token': RIOT_KEY})

for i in json.loads(summoner_games_data.data)["matches"]:
    Data_Game= url.request('GET',URL_SERVER+'/lol/match/v4/matches/'+str(i["gameId"]),headers={'X-Riot-Token': RIOT_KEY})
    a=pd.json_normalize(json.loads(Data_Game.data)["participantIdentities"])
    b=pd.json_normalize(json.loads(Data_Game.data)["participants"]).drop(["participantId"],axis=1)
    c=pd.json_normalize([json.loads(Data_Game.data)["teams"][0],json.loads(Data_Game.data)["teams"][0],json.loads(Data_Game.data)["teams"][0],json.loads(Data_Game.data)["teams"][0],json.loads(Data_Game.data)["teams"][0],json.loads(Data_Game.data)["teams"][1],json.loads(Data_Game.data)["teams"][1],json.loads(Data_Game.data)["teams"][1],json.loads(Data_Game.data)["teams"][1],json.loads(Data_Game.data)["teams"][1]]).drop(["bans","teamId"],axis=1)
    
    Data_Game=pd.concat([a,b,c],axis=1)
    if len(Total_data)== 0:
        Total_data= Data_Game
    else:
        Total_data=pd.concat([Total_data,Data_Game], axis=0,ignore_index=True)
    print(Total_data.shape)
    time.sleep(1)

Total_data.reset_index()
Total_data.to_csv ("/Users/milan/data_summoner_"+ SUMMONER_NAME +".txt", index = False, header=True)
```
